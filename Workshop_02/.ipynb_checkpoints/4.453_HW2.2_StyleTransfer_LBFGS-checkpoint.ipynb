# HW 2.2 Implementing Neural Style Transfer
In this notebook, you will implement neural style transfer as first proposed by [Gatys et al.](https://arxiv.org/abs/1508.06576) in 2015. The goal of neural style transfer is to apply a style filter to an image such that the new generated image retains the content of the original one while displaying style features of the filter.
By content, we mean the high-level structure of an image (e.g. buildings and landscape), while style refers to textures, colors, and patterns.  
In the example below (from this [blog post](https://medium.com/data-science-group-iitr/artistic-style-transfer-with-convolutional-neural-network-7ce2476039fd)), the content is a mountainous landscape, the style is "*The Scream*" by Munch, and the generated image draws the content of the landscape with textures from the style image.

```python
#@title Example
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

response = requests.get("https://cdn-images-1.medium.com/max/2000/1*uIlgYKjp-1ZboXK8ff6ztg.jpeg")
img = Image.open(BytesIO(response.content))

plt.axis("off")
plt.imshow(img)
plt.show()
```

While neural style transfer is not a particularly ground-breaking application of machine learning, it did have its 15 minutes of glory with apps like Prima and others that were all the rage on Instagram a few years back, and implementing neural style transfer is a great way to learn how to use neural networks for initially unintended purposes.  
As you may imagine, since we are dealing with images, we will use CNNs extensively. In fact, neural style transfer, and other image processing applications (e.g. [DeepDream](https://en.wikipedia.org/wiki/DeepDream)) were constructed based on our understanding of CNNs.  
As we saw in lecture 3, CNNs progressively process an image using convolutional and pooling filters. It has been shown that CNNs extract low-level image features (e.g. edges) in the bottom layers and progressively aggregate/summarize them into higher-level features (e.g. ears, eyes) in the top layers ([Zeiler & Fargus, 2013](https://arxiv.org/abs/1311.2901)).
In other words, the bottom layers extract features pertaining to the style of image whereas the top layers deal with the content of an image.  
Suppose the function *style(style_image)* returns the features extracted from the style image by the bottom layers as a flattened vector and the function *content(content_image)* returns the features extracted from the content image by the top layers.  
What do we want to achieve?


1.   We want our generated image *gen_image* to have a style as similar a possible as *style_image*. Mathematically speaking, this means we want ***style(gen_image)* and *style(style_image)* to be as similar as possible, which we will quantify using the squared distance between the two measures.**
2.  Similarly, we want our generated image *gen_image* to have a content as similar a possible as *content_image*. Again, this means ***content(gen_image)* and *content(content_image)* should be as close as possible.**

Suppose we initialize a random image *gen_image*. To turn this image into our styled image, we will optimize its pixel values such that they minimize the function  
**distance[ style( gen_image ), style( style_image ) ] + distance[ content( gen_image ), content( content_image ) ]**.

This makes sense conceptually, but how does this translate into code?  
  
  
First, let's load the images on which we'll like to use style transfer. Because we're downloading them from the web, the syntax is a bit intricated, but you can ignore that for now.

```python
import numpy as np
from PIL import Image
```

```python
# for some reason, this code does not work 100% of the time, rerun it a few times if it does not. Otherwise download images directly from url into your drive and load in google colab by mounting your drive
content_url = "https://adfastravel.com.au/wp-content/uploads/2018/08/ADFAS-Sardinia-and-Corsica.jpg"
style_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/db/Paul_Cezanne%2C_The_Gulf_of_Marseilles_Seen_from_L%27Estaque%2C_c1885.jpg/1280px-Paul_Cezanne%2C_The_Gulf_of_Marseilles_Seen_from_L%27Estaque%2C_c1885.jpg"
content_img = Image.open(requests.get(content_url, stream=True).raw)
style_img = Image.open(requests.get(style_url, stream=True).raw)

width, height= content_img.size
img_height = 500
img_width = int(width/height*img_height)

content_img = content_img.resize((img_width,img_height))
style_img = style_img.resize((img_width,img_height))

content_img = np.array(content_img)
style_img = np.array(style_img)

plt.axis("off")
plt.imshow(content_img)
plt.title('Content')
plt.show()
plt.axis("off")
plt.imshow(style_img)
plt.title('Style')
plt.show()


```

As mentioned above, neural style transfer uses a convolutional neural network as a means to evaluate style and content vectors. Here, we'll use VGG19, a variant of the VGG16 network seen in class. We import the model and preprocess the images to fit the input required by the network.

```python
from keras.applications import vgg19
content = vgg19.preprocess_input(np.expand_dims(content_img,axis=0))
style = vgg19.preprocess_input(np.expand_dims(style_img,axis=0))
```

Now, we need to setup a network that computes the layer activations for the style, content, and target images at the same time. Because the style and content images are constant, we use *K.constant*, and we use *K.placeholder* for the generated image since it will change over time.

```python
from keras import backend as K

content_image = K.constant(content)
style_image = K.constant(style)
gen_image = K.placeholder((1,img_height,img_width,3))
```

We want VGG19 to process the images simultaneously, so we'll combine the three images in a single batch. We then build VGG19 to take this batch as its input without including the top classification layer, which we do not care about for transfer learning.

```python
input_tensor = K.concatenate([content_image,style_image, gen_image], axis=0)
model = vgg19.VGG19(input_tensor=input_tensor, weights="imagenet", include_top = False)
```


Now, let's start defining the losses. 

### 1 [CODE REQUIRED]
First, **implement the content loss**. 

```python
def content_loss(content, generated):
    # return the sum of squared residuals between the content and generated vectors
    # hint: use backend K (we need to use K for the autodifferentiation to work!)
    return ((content - generated) ** 2).sum()/2
```

For the style loss, we actually compute the Gram matrix for each feature map (see [Gatys et al.](https://arxiv.org/abs/1508.06576) for details), which gives a map of the correlations in the original feature map.

```python
def gram_matrix(x):
    features = K.batch_flatten(K.permute_dimensions(x,(2,0,1)))
    gram = K.dot(features,K.transpose(features))
    return gram
```

### 2 [CODE REQUIRED]

Now, **implement the style loss** as defined in  [Gatys et al.](https://arxiv.org/abs/1508.06576). For the sake of simplicity replace $N_l$ and $M_l$ in the formula by, respectively, the number of channels (3) and the total pixel size ($width*height$) of the input image. Remember the *gram_matrix* function is already defined

```python
def style_loss(style, generated):
    Gl = gram_matrix(generated)
    Al = gram_matrix(style)
    El = (Gl - Al)**2/(4*style.shape[2] * (style.shape[0] * style.shape[1])^2)
    return El
```

We also add an optional third loss that only operates on the generated image to increase pixel continuity across the image. It computes the difference between neighbor pixels (horizontall and vertically).

```python
def variation_loss(generated):
    horizontal_diff = K.square(
    generated[:,:img_height-1,:img_width-1,:]-generated[:,1:,:img_width-1,:])
    vertical_diff = K.square(
    generated[:,:img_height-1,:img_width-1,:]-generated[:,:img_height-1,1::,:])
    return K.sum(K.pow(horizontal_diff+vertical_diff,1.25))
```

The loss that we'll minimize will be a weighted average of these three losses.  
We now need to define the final loss by connecting the losses defined above to their layers of concern.  
First, we devise a way to easily grab the layers we need:

```python
outputs_dict = dict([(layer.name, layer.output) for layer in model.layers])
content_layer = 'block5_conv2'
style_layers = ['block1_conv1','block2_conv1','block3_conv1','block4_conv1','block5_conv1']
```

Let's define the weights of each loss.

```python
variation_weight = 1e-4
style_weight = 1./len(style_layers)
content_weight = 0.025
```

Let's define the final loss. We define the loss is as a *variable* in the TensorFlow vocabulary because it depends on the image placeholder. We do not want to have to calculate the gradients ourselves, so we need to tell Tensorflow that the loss is dependent on other values for the library to build a graph, which will then be automatically differentiated.

```python
loss = K.variable(0.)
```

### 6 [NO ADDITIONAL CODE REQUIRED]

We have defined the loss and set up or graphs so we're all set to minimize the loss.
Let's define functions to evaluate the gradient and loss functions based on the current value of *gen_image*:

```python
grads = K.gradients(loss, gen_image)[0] # get gradient of loss with respect to gen_image

outputs = [loss]
if isinstance(grads, (list, tuple)):
    outputs += grads
else:
    outputs.append(grads)

f_outputs = K.function([gen_image], outputs)


def eval_loss_and_grads(x):
    if K.image_data_format() == 'channels_first':
        x = x.reshape((1, 3, img_height, img_width))
    else:
        x = x.reshape((1, img_height, img_width, 3))
    outs = f_outputs([x])
    loss_value = outs[0]
    if len(outs[1:]) == 1:
        grad_values = outs[1].flatten().astype('float64')
    else:
        grad_values = np.array(outs[1:]).flatten().astype('float64')
    return loss_value, grad_values
```

```python
# VGG19 automatically processes input images into a somewhat weird format, which entails that our generated image will have the same weird format
# as a result, we need to deprocess our generated image to be able to visualize it well.
def deprocess_img(x):
    x[:,:,0]+=103.939
    x[:,:,1]+=116.779
    x[:,:,2]+=123.68
    x=x[:,:,::-1]
    x = np.clip(x,0,255).astype('uint8')
    return x
```

```python
class Evaluator(object):

    def __init__(self):
        self.loss_value = None
        self.grads_values = None

    def loss(self, x):
        assert self.loss_value is None
        loss_value, grad_values = eval_loss_and_grads(x)
        self.loss_value = loss_value
        self.grad_values = grad_values
        return self.loss_value

    def grads(self, x):
        assert self.loss_value is not None
        grad_values = np.copy(self.grad_values)
        self.loss_value = None
        self.grad_values = None
        return grad_values

evaluator = Evaluator()
```

```python
# custom gradient descent implementation. We recommend using L-BFGS-B instead

def gradient_descent(x,iterations,step, decay=0.5, acceleration=1.1):
    last_loss=0
    for i in range(iterations):
        loss_and_grads = eval_loss_and_grads([x])
        loss, grad = loss_and_grads[0],loss_and_grads[1]
        print("Loss at iteration {0}:{1:.10e} | Current step size: {2:.4f}".format(i+1,loss,step))
        if loss > last_loss:
            step*=decay
        else:
            step*=acceleration
        #gradient descent step
        grad = grad/np.mean(np.abs(grad))
        x-=grad*step
        # plot at every iteration
        plt.axis("off")
        plt.imshow(deprocess_img(np.copy(x[0])))
        plt.show()
        last_loss = loss
    return x
```

Now, let's run the L-BFGS minimization algorithm to apply the styling. We'll initialize our generated image to the content image and run gradient descent. This way, we're sure we start at a *good* point. Do not be surprised if the optimization runs *very* slowly.

```python
from time import time
from scipy.optimize import fmin_l_bfgs_b
x = np.copy(content)

for i in range(20):
    print('Start of iteration', i)
    start_time = time()
    x, min_val, info = fmin_l_bfgs_b(evaluator.loss, x.flatten(),
                                     fprime=evaluator.grads, maxfun=20)
    print('Current loss value:', min_val)
    # save current generated image
   
    img = deprocess_img(x.copy().reshape(img_height, img_width, 3))
    plt.imshow(img)
    plt.axis("off")
    plt.show()
    end_time = time()
    print('Iteration %d completed in %ds' % (i, end_time - start_time))
```
