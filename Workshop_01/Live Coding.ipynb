```python
import math

def suqare_math(x):
    return int(math.pow(x, 2))

x = 6
y = suqare_math(x)
print(f"square of {x} is {y}")
```

```python
from random import randint

def 勾股定理(a, b):
    return math.sqrt(a*a + b*b)

a = randint(0,200)
b = randint(0,500)

斜边=勾股定理(a, b)
print(f"直角边是{a}和{b}的直角三角形斜边是{斜边}")
```
